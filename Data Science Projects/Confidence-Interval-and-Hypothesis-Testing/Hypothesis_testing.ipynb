# Hypothesis Testing
## Bombay Hospitality Ltd. — Weekly Operating Cost Analysis

**Background:**
Bombay Hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week is given by:

$$W = 1000 + 5X$$

where X = number of units produced in a week.

Restaurant owners claim their observed weekly operating costs are **higher** than what the model predicts.

```python
# Import required libraries
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

## Given Data

```python
# Given values
x_bar = 3050          # Sample mean weekly cost (Rs.)
mu_X = 600            # Mean units produced per week
sigma_X = 25          # Standard deviation of units produced
n = 25                # Sample size (number of restaurants)
alpha = 0.05          # Significance level

# Theoretical mean weekly cost from the model: W = 1000 + 5X
mu = 1000 + 5 * mu_X
print(f"Theoretical mean weekly cost (Mu): Rs. {mu}")

# Population standard deviation of W: σ_W = 5 * σ_X
sigma = 5 * sigma_X
print(f"Population standard deviation (Sigma): Rs. {sigma}")
print(f"Sample size (n): {n}")
print(f"Sample mean (x_Bar): Rs. {x_bar}")
```

## Step 1: State the Hypotheses

We are testing whether the weekly operating costs are **higher** than the model suggests.

```python
print("===== STEP 1: Hypotheses =====")
print(f"H₀ (Null Hypothesis)      : μ = Rs. {mu}  (costs follow the model)")
print(f"H₁ (Alternative Hypothesis): μ > Rs. {mu}  (costs are higher than the model)")
```

## Step 2: Calculate the Test Statistic

```python
print("===== STEP 2: Test Statistic =====")

# Standard error of the mean
SE = sigma / np.sqrt(n)
print(f"Standard Error (Sigma/np.sqrt(n)) = {sigma} / {np.sqrt(n)} = {SE}")

# Z-statistic
z_stat = (x_bar - mu) / SE
print(f"Z = {z_stat:.2f}")
```

## Step 3: Determine the Critical Value and P-value

- Reject H₀ if z > 1.645

```python
print("===== STEP 3: Critical Value & P-value =====")

# Critical value for one-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)
print(f"Critical value (z_α at α=0.05, right-tail): {z_critical:.2f}")

# P-value for right-tailed test
p_value = 1 - stats.norm.cdf(z_stat)
print(f"P-value (right-tail): {p_value:.2f}")
print(f"Significance level α: {alpha}")
```

## Step 4: Make a Decision

```python
print("===== STEP 4: Decision =====")
print(f"Calculated Z-statistic : {z_stat:.4f}")
print(f"Critical value (z_α)   : {z_critical:.4f}")
print(f"P-value                : {p_value:.6f}")
print(f"Alpha (α)              : {alpha}")
print()

if z_stat > z_critical:
    print("Decision: REJECT H₀")
    print(f"Reason: Z ({z_stat:.4f}) > Z_critical ({z_critical:.4f})")
else:
    print("Decision: FAIL TO REJECT H₀")
    print(f"Reason: Z ({z_stat:.4f}) ≤ Z_critical ({z_critical:.4f})")

if p_value < alpha:
    print(f"Also: p-value ({p_value:.6f}) < α ({alpha}) → Reject H₀")
else:
    print(f"Also: p-value ({p_value:.6f}) ≥ α ({alpha}) → Fail to Reject H₀")
```

## Step 5: Conclusion

```python
print("===== STEP 5: Conclusion =====")

if z_stat > z_critical:
    print("""
At a 5% significance level, we REJECT the null hypothesis.

There IS sufficient statistical evidence to support the restaurant owners' claim
that the weekly operating costs are HIGHER than what the theoretical model predicts.

The observed sample mean (Rs. 3,050) is significantly different from the
model-predicted mean (Rs. 4,000) in the direction claimed.
""")
else:
    print("""
At a 5% significance level, we FAIL TO REJECT the null hypothesis.

There is NOT sufficient statistical evidence to support the restaurant owners' claim
that the weekly operating costs are higher than the model suggests.
""")
```
