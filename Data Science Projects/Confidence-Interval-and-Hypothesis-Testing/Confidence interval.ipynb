```python
import numpy as np
import scipy.stats as stats

# Data provided in the scenario (in millions of characters)
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

# Basic statistics
n = len(data)
sample_mean = np.mean(data)
confidence_level = 0.99

print(f"Sample Size (n): {n}")
print(f"Sample Mean: {sample_mean:.4f} million characters")
```

```python
# Calculate sample standard deviation (ddof=1 for sample)
sample_std = np.std(data, ddof=1)

# Degrees of freedom for t-distribution
df = n - 1

# Calculate the t-critical value
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# Calculate the margin of error
margin_of_error_t = t_critical * (sample_std / np.sqrt(n))

# Calculate the confidence interval manually
lower_bound_t = sample_mean - margin_of_error_t
upper_bound_t = sample_mean + margin_of_error_t

# Alternatively, using scipy's built-in interval function
ci_t_scipy = stats.t.interval(confidence_level, df, loc=sample_mean, scale=sample_std/np.sqrt(n))

print("--- Task A: Using Sample Standard Deviation (t-distribution) ---")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"t-critical value: {t_critical:.4f}")
print(f"Margin of Error: {margin_of_error_t:.4f}")
print(f"99% Confidence Interval: ({lower_bound_t:.4f}, {upper_bound_t:.4f}) million characters")
```

```python
# Known population standard deviation
pop_std = 0.2

# Calculate the z-critical value for 99% confidence
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error_z = z_critical * (pop_std / np.sqrt(n))

# Calculate the confidence interval manually
lower_bound_z = sample_mean - margin_of_error_z
upper_bound_z = sample_mean + margin_of_error_z

# Alternatively, using scipy's built-in interval function
ci_z_scipy = stats.norm.interval(confidence_level, loc=sample_mean, scale=pop_std/np.sqrt(n))

print("--- Task B: Using Known Population Standard Deviation (z-distribution) ---")
print(f"Population Standard Deviation: {pop_std}")
print(f"z-critical value: {z_critical:.4f}")
print(f"Margin of Error: {margin_of_error_z:.4f}")
print(f"99% Confidence Interval: ({lower_bound_z:.4f}, {upper_bound_z:.4f}) million characters")
```
